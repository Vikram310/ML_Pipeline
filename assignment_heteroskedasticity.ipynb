### Heteroscedasticity

**Heteroskedasticity** can be very problematic with methods besides OLS (Ordinal Least Squares). For example, in logistic regression heteroskedasticity can produce biased and misleading parameter estimates.

Occurs more oftern in datasets that have large range between largest and smallest obsvered values.

The assumption of homoscedasticity (meaning “same variance”) is central to linear regression models.  Homoscedasticity describes a situation in which the error term (that is, the “noise” or random disturbance in the relationship between the independent variables and the dependent variable) is the same across all values of the independent variables.  Heteroscedasticity (the violation of homoscedasticity) is present when the size of the error term differs across values of an independent variable.

List Comprehension

Y = [Weight * x + (random.random() - 0.5) * Noise_Amplification * (Function(x)) + Y_Intercept for x in X]

where Function(x) would be something like f(x) = envelope_weight * x + envelope_intercept

```python
def sample (X,Y):
    fig = plt.figure(figsize=(10,6))
    #sns.regplot(X, Y, color="green", scatter=True,line_kws={"color": "red"})
    plt.scatter(X,Y)
    plt.title('This Is The Title')
    plt.xlabel('These Are The X Values')
    plt.ylabel('These Are The Y Values')
    plt.grid(color='r', linestyle='-', linewidth=0.2)
    plt.show()
    return

def Envelope (weight,x,intercept):
    env = weight * x + intercept
    return env
```

```python
Weight = 1                  # Constant, until Noise, then has correlation ?  Direction of correlation can create Positive, Negative , No Correlation
Noise_Amplification = 1     # Movement of Datapoint from Mean
Y_Intercept = 100           # Y axis start point
Envelope_weight = 1         # Increase in how far the spread can be. reduces the correlation coefficients
Envelope_Intercept = 0      # Larger value more spread the data points from X value, of the range of X

X = [x/10.0 for x in range(1000)]    # Python uses 0 Indexing, so range is 0-99
Y = [Weight * x + (random.random() -0.5) * Noise_Amplification * \
     Envelope(Envelope_weight,x,Envelope_Intercept) + Y_Intercept for x in X]

sample(X,Y)

from scipy.stats import pearsonr
# Pearson's correlation coefficient = covariance(X, Y) / (stdv(X) * stdv(Y))
corr,_ =pearsonr(X,Y)
print(corr)
```
